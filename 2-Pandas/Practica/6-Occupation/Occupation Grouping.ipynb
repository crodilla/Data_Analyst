### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user", sep="|")
users
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby("occupation")["age"].agg("mean")
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

#### Forma 1

```python
users_ocupacion = users.groupby("occupation")["gender"].count()
```

```python
hombres = users[users['gender'] == 'M']
hombres_ocupacion = hombres.groupby("occupation")["gender"].count()
```

```python
ratio_por_ocupacion = hombres_ocupacion / users_ocupacion
ratio_por_ocupacion.sort_values(ascending=False, inplace=True)
```

```python
ratio_por_ocupacion
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby("occupation")["age"].agg(["min", "max"])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['occupation', 'gender'])['age'].mean()
```

### Step 8.  For each occupation present the percentage of women and men

```python
datos_agrupados = pd.crosstab(index=users['occupation'], columns=users['gender'])
datos_agrupados
```

```python
datos_agrupados['Total'] = datos_agrupados['F'] + datos_agrupados['M']
datos_agrupados['% Hombres'] = 100 * datos_agrupados['M'] / datos_agrupados['Total']
datos_agrupados['% Mujeres'] = 100 * datos_agrupados['F'] / datos_agrupados['Total']
datos_agrupados
```
